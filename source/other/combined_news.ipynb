# create final dataset

```python
import pandas as pd

path_1 = "..//..//data//final_trusted_news.csv"
path_2 = "..//..//data//fake_news_dataset.csv"

df_1 = pd.read_csv(path_1)
df_1
```

```python
df_1 = df_1.drop(['Unnamed: 0'], axis=1)
df_1
```

```python
df_2 = pd.read_csv(path_2)
df_2
```

```python
df_combined = pd.concat(
    [df_1, df_2],
    ignore_index=True
)
df_combined
```

```python
df_combined.to_csv("..//..//data//final_dataset.csv")
```

```python
import os
import pandas as pd

# Шлях до директорії збереження
output_dir = "..//..//data//final_dataset"
base_filename = "trusted_news_part"
max_file_size_mb = 100
max_file_size_bytes = max_file_size_mb * 1024 * 1024

# Розділяємо на порції по рядках і записуємо
chunk_size = 15_000  # Стартова кількість рядків на спробу, можна підібрати
file_index = 1
start = 0

while start < len(df_combined):
    end = start + chunk_size
    temp_df = df_combined.iloc[start:end]

    output_path = os.path.join(output_dir, f"{base_filename}_{file_index}.csv")
    temp_df.to_csv(output_path, index=False, encoding='utf-8')

    # Перевірка розміру файлу
    file_size = os.path.getsize(output_path)

    # Якщо файл більший за ліміт, зменшуємо chunk_size і пробуємо знову
    while file_size > max_file_size_bytes and chunk_size > 1000:
        chunk_size = int(chunk_size * 0.9)
        end = start + chunk_size
        temp_df = df_combined.iloc[start:end]
        temp_df.to_csv(output_path, index=False, encoding='utf-8')
        file_size = os.path.getsize(output_path)

    print(f"[✔] Збережено {output_path} — {file_size / (1024*1024):.2f} МБ, {len(temp_df)} рядків")

    start += chunk_size
    file_index += 1
```
